```python
# prompt: linear regression code using python
# Submitted By: Seerat Fatima (BSCS 6B)
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term (intercept) to X
X_b = np.c_[np.ones((100, 1)), X]

# Calculate the parameters (weights) using the Normal Equation
# theta = (X_b^T * X_b)^(-1) * X_b^T * y
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the calculated parameters
print("Calculated parameters (theta):\n", theta)

# Make predictions using the calculated parameters
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add bias term
y_predict = X_new_b.dot(theta)

# Plot the results
plt.scatter(X, y, label='Training data')
plt.plot(X_new, y_predict, "r-", label='Linear regression prediction')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Example")
plt.legend()
plt.grid(True)
plt.show()
```
