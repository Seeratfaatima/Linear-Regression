```python
# prompt: Gradient descent code

import numpy as np

# Define the function to minimize
def objective_function(x):
  return x**2 - 4*x + 5

# Define the derivative of the function
def derivative(x):
  return 2*x - 4

# Set initial parameters
initial_x = 0  # Starting point
learning_rate = 0.1 # Step size
num_iterations = 100 # Number of steps to take

# Perform gradient descent
x = initial_x
for i in range(num_iterations):
  gradient = derivative(x)
  x = x - learning_rate * gradient
  print(f"Iteration {i+1}: x = {x}, function value = {objective_function(x)}")

print(f"\nOptimal x found: {x}")
print(f"Minimum function value: {objective_function(x)}")
```
